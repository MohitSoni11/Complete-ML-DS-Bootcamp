# Methods of Plotting

Firstly, there are 2 major ways of plotting -- using pyplot or using the object oriented technique. We are going to be using the object oriented technique because that allows for more flexibility in our graphs.

### Pyplot

```python
plt.plot(x, y)
plt.show()
```

### Object Oriented (Recommended)

```python
fig, ax = plt.subplots()
ax.plot(x, y) # adds data
plt.show()
```

# Creating Different Types of Plots

```python
x = np.linspace(0, 10, 100)
x[:10]
```

### Line plot

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 2); # adding the semi-colon means "plt.show()"
```

### Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar plot

```python
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}

fig, ax = plt.subplots()
bars = ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

# customizing graph
ax.set(title='Dan\'s Nut Butter Store',
       ylabel='Price ($)')

ax.bar_label(bars);
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values());
```

### Histogram

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x, bins=100);
```

### Subplots - 1st Method

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10,5))

# Plotting to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000), bins=30);
```

### Subplots - 2nd Method

```python
fig, ax =  plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10, 5))

# Plotting to each different axis
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000), bins=30);
```

# Plotting from Pandas Dataframes

# Plotting Data from another Dataset

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head()
```

```python
# Seeing the distribution of the age column
heart_disease['age'].plot.hist(bins=40);
```

```python
# Not ideal because we want different scales for each plot
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

# Using the Object-Oriented (OO) Plotting Method with DataFrames

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# This is not the OO method, it is the pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(10, 6));
```

```python
# OO method
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10, 10), sharex=True)

scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
scatter2 = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

# Customize the plot
ax0.set(title='Heart Disease Diagnosis based on Cholesterol and Age', ylabel='Cholesterol')
ax1.set(title='Heart Disease Diagnosis based on Max Heart Rate and Age', xlabel='Age',ylabel='Max Heart Rate')

# Adding a legend
ax0.legend(*scatter.legend_elements(), title='Target')
ax1.legend(*scatter2.legend_elements(), title='Target')

# Add a horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle='--')
ax1.axhline(over_50['thalach'].mean(), linestyle='--')

# Adding title to entire figure
fig.suptitle('Heart Disease Analysis', fontsize=20, fontweight='bold');
```
